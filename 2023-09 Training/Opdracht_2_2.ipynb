```python
import numpy as np
import matplotlib.pyplot as plt
```

### Linear model

```python
a = 2
b = 5

model = lambda x: a * x + b
```

```python
x_min = 0
x_max = 10
n = 80
```

```python
x_model = np.linspace(x_min, x_max, n)
y_model = model(x_model)
```

```python
x_data = np.random.uniform(x_min, x_max, n)
```

```python
spread = 3
noise = np.random.randn(n) * spread
```

```python
y_data = model(x_data) + noise
```

```python
# plt.plot(x_model, y_model, color='red', linewidth=1, alpha=0.8)
plt.scatter(x_data, y_data)
plt.show()
```

### Linear Regression

```python
coefs = np.polyfit(x_data, y_data, deg = 1)
coefs
```

```python
model = np.poly1d(coefs)
```

```python
y_predicted = model(x_data)
```

```python
plt.scatter(x_data, y_data)
plt.plot(x_model, y_model, color='green', linewidth=1, alpha=0.8)
plt.scatter(x_data, y_predicted, color = 'red', s = 2)
plt.show()
```

## Other models

```python
a = 8
b = 3
c = 4

model = lambda x: a * np.sin(c * x) + b

plt.plot(x_model, model(x_model), color='red', linewidth=1, alpha=0.8)
plt.scatter(x_data, model(x_data) + noise)
plt.show()
```

```python
a = 1
b = -8
c = 30

model = lambda x: a*x**2 + b*x + c

plt.plot(x_model, model(x_model), color='red', linewidth=1, alpha=0.8)
plt.scatter(x_data, model(x_data) + noise)
plt.show()
```

```python
a = 0.05
b = 2
c = 0

model = lambda x: a*b**x + c

plt.plot(x_model, model(x_model), color='red', linewidth=1, alpha=0.8)
plt.scatter(x_data, model(x_data) + noise)
plt.show()
```
