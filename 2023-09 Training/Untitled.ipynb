```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.random.randn(1000)
y = np.random.randn(1000)
```

```python
bins = np.arange(-4, 4.01, 0.5)
plt.hist(x, bins = bins, orientation='horizontal')
plt.show()
```

```python
def maak_hist(data, stepsize = 0.5):
    bins = np.arange(np.min(data), np.max(data) + 0.01, stepsize)
    plt.hist(x, bins = bins, orientation='horizontal', color = 'darkred')
    plt.show()
```

```python
maak_hist(x)
```

```python
sns.set_theme(style="darkgrid")
sns.jointplot(x=x, y=y)
plt.show()
```

```python
print(plt.style.available)
```

```python
plt.scatter(x, y, color = 'green', marker = 's', alpha = 0.3)
plt.title('Demo scatterplot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['data'])
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, sharey = True, figsize = (12, 6))

ax[0].scatter(x, y, color = 'indigo', marker = 's', alpha = 0.3)
ax[0].set_title('Demo scatterplot')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].legend(['data'])
ax[0].annotate('LET OP!', xy = (-3, 2), xytext=(-4, 3), arrowprops = {'arrowstyle':'->'})



ax[1].scatter(y, x, color = 'k', marker = '^', s = 100, alpha = 0.3)
ax[1].set_title('Another Demo scatterplot')
ax[1].set_xlabel('y')
ax[1].legend(['data'])

plt.show()
```

```python
plt.plot(np.sort(x), np.sort(y), lw = 3, ls = 'dashed')
plt.plot(np.sort(x), np.sort(y/2), lw = 3, ls = 'dashed')
plt.plot(np.sort(x), np.sort(y/3), lw = 3, ls = 'dashed')
plt.plot(np.sort(x), np.sort(y/4), lw = 3, ls = 'dashed', color = '0.1')

plt.show()
```
