### Regression mpg vs. cylinders, displacement, horsepower, weight

```python
target_name = 'mpg'
feature_names = ['cylinders', 'displacement', 'horsepower', 'weight']
```

```python
y = df[target_name].values
```

```python
X = df[feature_names].values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()

regressor.fit(X, y)
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

```python
y_predicted = regressor.predict(X)
```

```python
X[0]
```

```python
y[0]
```

```python
y_predicted[0]
```

```python
(regressor.coef_ * X[0]).sum() + regressor.intercept_
```

```python
plt.hist(y_predicted - y)
plt.show()
```

```python
plt.hist((y_predicted - y) / y * 100)
plt.title('Relatieve verschil in %')
plt.xlabel('%')
plt.show()
```

```python
plt.scatter(y, y_predicted)
plt.title('MPG vs Predicted MPG')
plt.xlabel('MPG')
plt.ylabel('Predicted MPG')
plt.show()
```

### R-squared metric

```python
from sklearn.metrics import r2_score

r_squared = r2_score(y, y_predicted)
r_squared
```

### Nearest Neighbor Regression

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
regressor = KNeighborsRegressor(n_neighbors=5)

regressor.fit(X, y)

y_predicted = regressor.predict(X)
```

```python
from sklearn.metrics import r2_score

r_squared = r2_score(y, y_predicted)
r_squared
```

```python
plt.scatter(y, y_predicted)
plt.title('MPG vs Predicted MPG')
plt.xlabel('MPG')
plt.ylabel('Predicted MPG')
plt.show()
```
