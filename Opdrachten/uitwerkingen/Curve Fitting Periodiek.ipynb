```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
a = 2
b = 5
c = 2
d = 2

model = lambda x: a * np.sin(c*x + d) + b
```

```python
n = 100
x_min = 0
x_max = 10
spread = 1

x_model = np.linspace(x_min, x_max, n)
y_model = model(x_model) 

ruis = np.random.randn(n) * spread

x_data = np.random.uniform(x_min, x_max, n)
y_data = model(x_data) + ruis

# plt.plot(x_model, y_model, color='r', linewidth=1)
plt.scatter(x_data, y_data)

plt.show()
```

```python
plt.plot(x_model, y_model, color='r', linewidth=1)
plt.axhline(0)
plt.axvline(0)
```

```python
model_to_optimize = lambda x, a, b, c, d: a * np.sin(c*x + d) + b

bounds = [(1,4,1,1),(3,6,3,3)]

parameters, *_ = scipy.optimize.curve_fit(model_to_optimize, x_data, y_data, bounds=bounds)

print(parameters)
```
