# Opdracht 2.2 - Ruis toevoegen aan een model

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 2
b = 5

model = lambda x: a * x + b

# or

def model(x):
    return a * x + b
```

```python
x_min = 0
x_max = 10

n = 100
```

```python
x_model = np.linspace(x_min, x_max, n)
y_model = model(x_model)
```

```python
plt.grid()
plt.plot(x_model, y_model)
plt.axhline(0, c='k')
plt.axvline(0, c='k')
plt.title('y = 2x + 5')
plt.show()
```

```python
x_data = np.random.uniform(x_min, x_max, n)
```

```python
spread = 3

ruis = np.random.randn(n) * spread

# or

ruis = np.random.normal(0, spread, n)
```

```python
y_data = model(x_data) + ruis
```

```python
plt.scatter(x_data, y_data)

plt.show()
```

```python
plt.scatter(x_data, y_data)
plt.show()
```

```python
plt.grid()
plt.plot(x_model, y_model, color='red', linewidth=1)
plt.scatter(x_data, y_data)
plt.axhline(0, c='k')
plt.axvline(0, c='k')
plt.title('y = 2x + 5')
plt.show()
```

```python
np.corrcoef(x_data, y_data)
```

###  Sinusvormig

```python
a = 2
b = 5
c = 2

model = lambda x: a * np.sin(c*x) + b

n = 100
x_min = 0
x_max = 10
spread = 1

x_model = np.linspace(x_min, x_max, n)
y_model = model(x_model) 

ruis = np.random.randn(n) * spread

x_data = np.random.uniform(x_min, x_max, n)
y_data = model(x_data) + ruis

# plt.plot(x_model, y_model, color='r', linewidth=1)
plt.scatter(x_data, y_data)

plt.show()
```

### Kwadratisch

```python
a = 0.1
b = 0.2
c = 0

model = lambda x: a*x**2 + b*x + c

n = 100
x_min = -10
x_max = 10
spread = 1

x_model = np.linspace(x_min, x_max, n)
y_model = model(x_model) 

ruis = np.random.randn(n) * spread

x_data = np.random.uniform(x_min, x_max, n)
y_data = model(x_data) + ruis

# plt.plot(x_model, y_model, color='r', linewidth=1)
plt.scatter(x_data, y_data)

plt.show()
```

### Exponentieel

```python
a = 0.1
b = 2
c = 0

model = lambda x: a*b**x + c

n = 100
x_min = 0
x_max = 8
spread = 5

x_model = np.linspace(x_min, x_max, n)
y_model = model(x_model) 

ruis = np.random.randn(n) * spread

x_data = np.random.uniform(x_min, x_max, n)
y_data = model(x_data) + ruis

plt.plot(x_model, y_model, color='r', linewidth=1)
plt.scatter(x_data, y_data)

plt.show()
```
