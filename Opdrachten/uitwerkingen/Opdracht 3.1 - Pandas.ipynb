### Voeg een nieuwe kolom toe

```python
df['name'] = df['first_name'] + ' ' + df['last_name']
```

#### alternatives

```python
df['name'] = df['first_name'].str.cat(df['last_name'], sep=' ')
df['name'] = df.assign(name = df['first_name'] + ' ' + df['last_name'])
df['name'] = df.assign(name = ' '.join(df[['first_name','last_name']]))
df['name'] = df.assign(name = lambda row: ' '.join(row[['first_name','last_name']]))
df['name'] = df.apply(lambda row: row['first_name'] + ' ' + row['last_name'], axis=1)
df['name'] = df.apply(lambda row: ' '.join(row[['first_name','last_name']]), axis=1)
df['name'] = df[['first_name','last_name']].apply(lambda x:' '.join(x), axis=1)
```

#### timeit

```python
%timeit df['name'] = df['first_name'] + ' ' + df['last_name']

%timeit df['first_name'].str.cat(df['last_name'], sep=' ')
%timeit df.assign(name = df['first_name'] + ' ' + df['last_name'])
%timeit df.assign(name = ' '.join(df[['first_name','last_name']]))
%timeit df.assign(name = lambda row: ' '.join(row[['first_name','last_name']]))
%timeit df.apply(lambda row: row['first_name'] + ' ' + row['last_name'], axis=1)
%timeit df.apply(lambda row: ' '.join(row[['first_name','last_name']]), axis=1)
%timeit df[['first_name','last_name']].apply(lambda x:' '.join(x), axis=1)
```

### Filter rows

```python
df[df['city']=='Montreal']

df.query('city=="Montreal"')
```

```python
columns = ['first_name','last_name','name','city']

df.loc[df['city'].isin(['Montreal', 'Vancouver']), columns]
```

#### alternatives

```python
df.loc[df['city'].isin(['Montreal', 'Vancouver']), columns]
df.loc[(df['city']=='Montreal') | (df['city']=='Vancouver'), columns]
df.loc[np.logical_or(df['city']=='Montreal', df['city']=='Vancouver'), columns]
df.loc[df['city'].str.match('Montreal|Vancouver'), columns]
```

#### timeit

```python
%timeit df.loc[(df['city']=='Montreal') | (df['city']=='Vancouver'), columns]
%timeit df.loc[np.logical_or(df['city']=='Montreal', df['city']=='Vancouver'), columns]
%timeit df.loc[df['city'].isin(['Montreal', 'Vancouver']), columns]
%timeit df.loc[df['city'].str.match('Montreal|Vancouver'), columns]
```

### Multiple filter criteria

### Convert data types

```python
df_temp['name'] = df_temp['name'].astype('string')
df_temp['email'] = df_temp['email'].astype('string')
df_temp['city'] = df_temp['city'].astype('category')
df_temp.dtypes
```
