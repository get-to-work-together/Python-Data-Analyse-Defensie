# Opdracht 4.2 - Verschillende plots maken

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
filename = '../datasets/iris.csv'

df = pd.read_csv(filename)

df.head()
```

```python
print(df.shape)
print(df.dtypes)
```

```python
list(df['species'].unique())
```

```python
species_count = df['species'].value_counts()
species_count
```

```python
# species_count.plot.bar()
# species_count.plot(kind='bar')

plt.bar(species_count.index, species_count, width=0.4)

plt.show()
```

```python
# species_count.plot.barh()

plt.barh(species_count.index, species_count, height=0.5, label='x')
plt.xlim(0, 60)

for i, v in enumerate(species_count):
    plt.text(v + 1, i-0.04, str(v), color='k')
```

```python
ax = species_count.plot(kind='bar')

ax.set_title('Aantal per Iris soort')
ax.set_xlabel('Iris soort')
ax.set_ylabel('Aantal')

plt.show()
```

```python
colors = dict(zip(df['species'].unique(), ['#028482','#7ABA7A','#B76EB8']))
colors
```

```python
ax = species_count.plot(kind='barh', color=colors.values())

ax.set_title('Aantal per Iris soort')
ax.set_xlabel('Aantal')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_color('#888888')
ax.spines['bottom'].set_linewidth(1)
ax.spines['left'].set_color('#888888')
ax.spines['left'].set_linewidth(1)

plt.show()
```

```python
fig, axs = plt.subplots(1, 4, figsize=(14, 3))

axs[0].hist(df['sepal_length'])
axs[0].set_title('sepal_length')

axs[1].hist(df['sepal_width'])
axs[1].set_title('sepal_width')

axs[2].hist(df['petal_length'])
axs[2].set_title('petal_length')

axs[3].hist(df['petal_width'])
axs[3].set_title('petal_width')

plt.show()
```

```python
fig, axs = plt.subplots(1, 4, figsize=(14, 3))

def maak_hist(df, columnname, ax):
    df[columnname].hist(ax = ax, grid=False)
    ax.set_title(columnname)

maak_hist(df, 'sepal_length', ax=axs[0])
maak_hist(df, 'sepal_width', ax=axs[1])
maak_hist(df, 'petal_length', ax=axs[2])
maak_hist(df, 'petal_width', ax=axs[3])

plt.show()
```

```python
plt.figure(figsize=(14, 3))

plt.subplot(1, 4, 1, title = 'sepal_length')
df['sepal_length'].hist()
plt.xlim(0, 10)

plt.subplot(1, 4, 2, title = 'sepal_width')
df['sepal_width'].hist()
plt.xlim(0, 10)

plt.subplot(1, 4, 3, title = 'petal_length')
df['petal_length'].hist()
plt.xlim(0, 10)

plt.subplot(144, title = 'petal_width')
df['petal_width'].hist()
plt.xlim(0, 10)

plt.show()
```

```python
# MATLAB style

plt.figure(figsize=(14, 3))

plt.subplot(1,4,1)
plt.title('sepal_length')
plt.hist(df['sepal_length'])

plt.subplot(1,4,2)
plt.title('sepal_width')
plt.hist(df['sepal_width'])

plt.subplot(1,4,3)
plt.title('petal_length')
plt.hist(df['petal_length'])

plt.subplot(1,4,4)
plt.title('petal_width')
plt.hist(df['petal_width'])

plt.show()
```

```python
# Object oriented style

fig, ax = plt.subplots(1, 4, figsize=(14, 3), sharey=True)

ax[0].set_title('sepal_length')
ax[0].hist(df['sepal_length'])

ax[1].set_title('sepal_width')
ax[1].hist(df['sepal_width'])

ax[2].set_title('petal_length')
ax[2].hist(df['petal_length'])

ax[3].set_title('petal_width')
ax[3].hist(df['petal_width'])

plt.show()
```

```python
fig, ax = plt.subplots(1, 4, figsize=(14, 3), sharey=True)

ax[0].set_title('sepal_length')
ax[0].hist('sepal_length', data=df)

ax[1].set_title('sepal_width')
ax[1].hist('sepal_width', data=df)

ax[2].set_title('petal_length')
ax[2].hist('petal_length', data=df)

ax[3].set_title('petal_width')
ax[3].hist('petal_width', data=df)

plt.show()
```
