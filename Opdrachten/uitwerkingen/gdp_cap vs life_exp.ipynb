### gdp_cap vs life_exp

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename = '../datasets/wb_gdp_cap_vs_life_exp.csv'
df = pd.read_csv(filename)
```

```python
df
```

```python
df.sort_values('population')
```

```python
df = df[df['cont']=='Europe']
```

```python
gdp_cap = df['gdp_cap']
life_exp = df['life_exp']
```

```python
size = df['population'] * 0.000007
```

```python
continent_to_color_mapping = {
    'Asia':'red',
    'Europe':'limegreen',
    'Africa':'deepskyblue',
    'Americas':'gold',
    'Oceania':'black'
}

col = df['cont'].map(continent_to_color_mapping)
```

```python
# Figure
plt.figure(figsize = (14, 16))
ax = plt.gca()
ax.set_facecolor('whitesmoke')

# Scatter plot
plt.scatter(x = gdp_cap, 
            y = life_exp, 
            s = size, 
            c = col, 
            alpha = 0.3)

# Previous customizations
plt.xscale('log') 
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks([1000,10000,100000], ['1k','10k','100k'])

# Additional customizations
# plt.text(2452, 64.7, 'India')
# plt.text(4959, 73.0, 'China')
# plt.annotate('India', xy=(2452, 64.7), xytext=(3000, 70) ,arrowprops={'arrowstyle':'->'})
# plt.annotate('China', xy=(4959, 73.0), xytext=(5000, 80) ,arrowprops={'arrowstyle':'->'})


# Annotate with labels
for i, txt in enumerate(df['country']):
    plt.annotate(txt, 
                 xy = (gdp_cap.iat[i], life_exp.iat[i]),
                 xytext=(5, 2), #show annotate
                 textcoords='offset points',
                 ha='left',
                 va='top',
                 alpha=0.7
                )


# Add grid() call
plt.grid(True)

# Show the plot
plt.show()
```
