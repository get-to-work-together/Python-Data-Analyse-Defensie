### Data Wrangling & Cleaning

```python
df.info()
```

```python
df['origin'] = df['origin'].replace({1:'USA', 2:'Europe', 3:'Asia'}).astype('category')
```

```python
df['horsepower'] = df['horsepower'].replace('?', np.NaN).astype('float')
```

```python
df.rename(columns={'%mpg': 'mpg'}, inplace=True)
```

```python
df[df['horsepower'].isna()]
```

```python
df.dropna(inplace = True)
```

```python
df['car name'] = df['car name'].astype('string')
```

```python
df[['make', 'model']] = df['car name'].str.split(n = 1, expand = True)
```

```python
df['make'].value_counts().sort_index()
```

```python
typos = {
    'chevroelt': 'chevrolet',
    'chevy': 'chevrolet',
    'maxda': 'mazda',
    'mercedes-benz': 'mercedes',
    'toyouta': 'toyota',
    'vokswagen': 'volkswagen',
    'vw': 'volkswagen'
}
df['make'] = df['make'].replace(typos)
```
