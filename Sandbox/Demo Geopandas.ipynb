```python
land = nederland[nederland['H2O']=='NEE'].copy()
```

```python
land.plot(figsize=(12, 12))
```

```python
land.plot(column='BEV_DICHTH', figsize=(12, 12))
```

```python
land[['GM_CODE','GM_NAAM','BEV_DICHTH','AANT_INW','OPP_LAND']].sort_values('BEV_DICHTH', ascending=False).tail(10)
```
