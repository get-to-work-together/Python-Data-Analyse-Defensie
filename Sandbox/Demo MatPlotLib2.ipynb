# MatPlotLib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0.1, 30, 100)
```

```python
y1 = np.sin(x)/x
y2 = np.cos(x)/x
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y2 + 1)

plt.ylim(top = 2)

plt.axhline(0, color='black', linewidth=1, alpha=0.5)
plt.axhline(1, color='red', linewidth=1, linestyle='dashed')

plt.title('X versus Y')
plt.xlabel('X waarden')
plt.ylabel('Y waarden')

plt.annotate('important', xy=(10,1), xytext=(10, 1.5), arrowprops={'arrowstyle':'->'})

plt.legend(['y1','y2','y2 + 1'])

plt.show()
```

```python
x1 = np.random.normal(3, 2, 100)
x2 = np.random.normal(30, 10, 100)
```

```python
plt.scatter(x1+1, x2+10, marker='+', color='blue', alpha=0.8)
plt.scatter(x1, x2, marker='d', color='red', alpha=0.8

plt.xlim(-6, 10)
plt.ylim(-10, 70)

plt.title('X1 versus X2')
plt.xlabel('X1')
plt.ylabel('X2')

plt.axhline(30, linestyle='dotted', color='green')
plt.axvline(3, linestyle='dotted', color='green')

plt.show()
```

```python
plt.hist(x1, bins=20)

plt.title('histogram')
plt.xlabel('x1')
plt.ylabel('aantal')
plt.axvline(3, c='r', ls='dotted')

plt.show()
```

```python
fig, axs = plt.subplots(1, 2, sharey=True, figsize=(14, 3))

axs[0].hist(x1, bins=20, color='#DB1F48')
axs[1].hist(x2, bins=20)

plt.show()
```

```python
plt.pie([2,5,5])

plt.show()
```

```python
plt.boxplot([x1, x2])

plt.show()

```

```python
plt.violinplot([x1, x2])

plt.show()
```

```python
plt.polar(x, np.sin(x))
plt.polar(x, np.cos(x))
plt.show()
```

```python
import matplotlib.pyplot as plt


labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
men_std = [2, 3, 4, 1, 2]
women_std = [3, 5, 2, 3, 3]
width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(labels, men_means, width, yerr=men_std, label='Men')
ax.bar(labels, women_means, width, yerr=women_std, bottom=men_means,
       label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()

plt.show()
```

```python
import numpy as np
np.random.seed(19680802)
import matplotlib.pyplot as plt


fig, ax = plt.subplots()
for color in ['tab:blue', 'tab:orange', 'tab:green']:
    n = 750
    x, y = np.random.rand(2, n)
    scale = 400.0 * np.random.rand(n)
    ax.scatter(x, y, c=color, s=scale, label=color,
               alpha=0.3, edgecolors='none')

ax.legend()
ax.grid(True)

plt.show()
```
