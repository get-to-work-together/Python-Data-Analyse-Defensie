## Opdracht 2.4 - lineaire regressie

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### definieer een lineaire functie

```python
def f(x):
    a = 5
    b = 10
    return a * x + b
```

```python
x = np.linspace(1, 10, 101)
```

```python
plt.plot(x, f(x))
plt.show()
```

```python
spreiding = 15
ruis = np.random.normal(0, spreiding, 101)
y = f(x) + ruis
```

```python
plt.plot(x, f(x), color='red')
plt.scatter(x, y)
plt.show()
```

### Fitting

```python
coefs = np.polyfit(x, y, 1)
model = np.poly1d(coefs)
```

```python
coefs
```

```python
plt.plot(x, f(x), color='red')
plt.plot(x, model(x), color='green')
plt.scatter(x, y)
```
