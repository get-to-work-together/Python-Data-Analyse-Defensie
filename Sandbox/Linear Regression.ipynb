# Relation between bill length and flipper length

```python
x = df['bill_length_mm'].values
y = df['flipper_length_mm'].values
```

```python
plt.scatter(x, y)

plt.show()
```

```python
np.corrcoef(x, y)
```

```python
degree = 1
coefs = np.polyfit(x, y, degree)
```

```python
coefs
```

```python
regression = np.poly1d(coefs)
```

```python
plt.scatter(x, y)
plt.plot(x, regression(x), 'red')

plt.show()
```
