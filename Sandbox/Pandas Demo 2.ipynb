## Get data

```python
url = 'https://nl.wikipedia.org/wiki/Provincies_van_Nederland'

df = pd.read_html(url, decimal=',', thousands='.')[0]
```

```python
df
```

```python
df['Provincie'] = df['Provincie'].astype('string')
df['Hoofdstad'] = df['Hoofdstad'].astype('string')
df['Gemeente met demeeste inwoners'] = df['Gemeente met demeeste inwoners'].astype('string')
```

```python
df.dtypes
```

## Indexing and slicing

```python
df
```

```python
df[['Provincie']]
```

```python
df[['Provincie','Hoofdstad','Bevolking[1]']]
```

```python
df['Adressering'] = df['Hoofdstad'] + ', ' + df['Provincie']
```

```python
df
```

```python
df['inw per km2'] =(df['Bevolking[1]']/df['Oppervlakte(land) in km²']).round().astype('int')
```

```python
df[['inw./km²[2]','inw per km2']]
```

```python
df[df['Bevolking[1]']>1000000]
```

```python
df.loc[df['Bevolking[1]']>1000000, ['Provincie','Hoofdstad','Bevolking[1]']]
```
