# Data Cleaning

```python
df.dropna(inplace=True)
```

```python
target_column = 'sex'
feature_columns = ['species','island','bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']
# feature_columns = list(df.columns[1:])
```

# One-hot Encoding

```python
df['island'].unique()
```

```python
df['species'].unique()
```

```python
df_dummies = pd.get_dummies(df[['island','species']], drop_first=True, dtype = 'int')
df_dummies
```

# Scaling

```python
numeric_features = ['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(df[numeric_features])
```

```python
scaled = scaler.transform(df[numeric_features])
scaled
```

# Features

```python
features = df_dummies.copy()
features[numeric_features] = scaled
```

```python
features
```

```python
features.shape
```

# Target

```python
target = df['sex']
```

```python
target.shape
```

# Split in train en test datasets

```python
from sklearn.model_selection import train_test_split
```

```python
# np.random.seed(1234)

df_train_features, \
df_test_features, \
df_train_target, \
df_test_target = train_test_split(features, target, test_size = 0.3)
```

```python
print('df_train_features', df_train_features.shape)
print('df_train_target', df_train_target.shape)
print('df_test_features', df_test_features.shape)
print('df_test_target', df_test_target.shape)
```

# Classification

```python
df_results = pd.DataFrame(df_test_target)
```

## k-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors = 3)

classifier.fit(df_train_features, df_train_target)
```

```python
df_results['3NN_predicted'] = classifier.predict(df_test_features)
df_results['3NN_correct'] = df_results['3NN_predicted'] == df_results['sex']
```

```python
df_results
```

```python
df_results['3NN_correct'].value_counts()
```

```python
import sklearn.metrics as sklm

accuracy = sklm.accuracy_score(df_results['sex'], df_results['3NN_predicted'])
accuracy
```
