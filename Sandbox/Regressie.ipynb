```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
```

## Data genereren

```python
model = lambda x: 10 * x + 60
```

```python
n = 100
x = np.linspace(0, 20, n)
y = model(x)
```

```python
plt.plot(x, y)
plt.ylim(bottom = 0)
plt.show()
```

```python
ruis = np.random.randn(len(x)) * 50
```

```python
y = model(x) + ruis
```

```python
plt.plot(x, model(x), alpha=0.4)
plt.scatter(x, y, marker='+', color='green')
plt.show()
```

```python
coefs = np.polyfit(x, y, 1)
coefs
```

```python
# fit = lambda x: 9.80843611 * x + 63.01093377
fit = np.poly1d(coefs)
```

```python
plt.plot(x, model(x), alpha=0.4)
plt.scatter(x, y, marker='+', color='green')
plt.plot(x, fit(x), color='red')
plt.show()
```

```python
regression = stats.linregress(x, y)
slope, intercept, *_ = regression
```

```python
fit = lambda x: slope * x + intercept
```

```python
plt.plot(x, model(x), alpha=0.4)
plt.scatter(x, y, marker='+', color='green')
plt.plot(x, fit(x), color='red')
plt.show()
```

```python
regression
```

```python
errors = y - fit(x)
errors
```

```python
plt.hist(errors)
plt.show()
```

```python
np.mean(np.abs(errors))     # Mean Absolute Error]
```

```python
np.sqrt(np.mean(errors ** 2))     # Mean Sqare Root Errors
```
