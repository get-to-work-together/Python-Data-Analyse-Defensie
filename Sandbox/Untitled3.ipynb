```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(0.1, 15, 100)
```

```python
y = np.sin(2*x) / x
```

```python
plt.figure(figsize=(14, 5))

for i in range(10):
    plt.plot(x, i * y)

plt.legend(['A','B','C','D','E','F','G','H','I','J'])

ax = plt.gca()
ax.set_facecolor("orange")

plt.title('aantal versus meters')
plt.xlabel('meters')
plt.ylabel('aantal')

plt.show()
```

```python
fig, ax = plt.subplots(2, 3, sharex=True, sharey=True, figsize=(14, 6))

ax[0,0].plot(x, x*2/30)
ax[0,0].set_title('A')
ax[0,1].plot(x, x**2/200)
ax[0,1].set_title('B')
ax[0,2].plot(x, np.sin(x))
ax[0,2].set_title('C')
ax[1,0].plot(x, np.log(x))
ax[1,0].set_title('Nog en A')
ax[1,1].plot(x, x**5/600000)
ax[1,1].set_title('Bla Bla')
ax[1,2].plot(x, 1/x/10)
ax[1,2].set_title('CCCCCC')

plt.show()
```

```python
x = np.linspace(1, 10, 11)

y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, linewidth=1, marker='s')
plt.plot(x, y2, linewidth=1, marker='s')

plt.show()
```

```python
x = np.random.normal(5, 10, 1000)
y1 = np.random.normal(2, 3, 1000)
y2 = np.random.normal(10, 5, 1000)

plt.scatter(x, y1, marker='o', alpha=0.2, color='green')
plt.scatter(x, y2, marker='o', alpha=0.2, color='blue')

plt.show()
```

```python
bins = np.linspace(-40, 40, 17)

plt.hist(x, bins=bins, color='lightblue')

plt.axvline(5, color='red', linestyle='-.')
plt.axhline(100, color='green', linewidth=20, alpha=0.3)

plt.show()
```

```python
plt.barh(penguins['species'], penguins['body_mass_g'])

plt.ylabel('species')
plt.xlabel('body mass [g]')


plt.show()
```

```python
kenmerken = ['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']

fig, ax = plt.subplots(1, 4, figsize=(14,4))

for i, kenmerk in enumerate(kenmerken):
    ax[i].bar(penguins['species'], penguins[kenmerk])
    ax[i].set_title(kenmerk)
        
plt.show()
```

```python
colors = {'Adelie':'#870A30','Chinstrap':'#EC8FD0','Gentoo':'#D43790'}
```

```python
plt.grid(alpha=0.5)

species = colors.keys()
for specie in species:
    color = colors[specie]
    data = penguins[penguins['species']==specie]
    plt.scatter(data['bill_length_mm'], data['bill_depth_mm'], color=color, marker='o')

plt.xlabel('bill_length_mm')
plt.ylabel('bill_depth_mm')
# plt.legend(species)

plt.text(47, 16, 'X', c='b', font = {'weight': 'bold','size': 22})
# plt.annotate('Important', xy=(53, 19),  xytext = (60, 18), arrowprops={'arrowstyle':'->'})

plt.show()

```
