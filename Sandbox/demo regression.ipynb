```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = 'data.csv'

data = np.loadtxt(filename, delimiter = ';', skiprows = 1)
data
```

```python
data.shape
```

```python
data = data * 2.54
```

```python
fathers = data[:, 0]
sons = data[:, 1]
```

```python
plt.scatter(fathers, sons)
plt.show()
```

```python
coefs = np.polyfit(fathers, sons, deg = 1)
```

```python
model = np.poly1d(coefs)
```

```python
predicted_sons = model(fathers)
```

```python
plt.scatter(fathers, sons)
plt.scatter(fathers, predicted_sons, color = 'red', s = 1)
plt.show()
```

```python
model(180)
```
