## Demo MatPlotLib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1, 10, 500)
y1 = np.sin(x**2)
y2 = np.cos((x-3)**2)
```

```python
plt.figure(figsize = (12, 3))
plt.plot(x, y1, color = 'darkmagenta', linewidth=1.5, alpha=0.7)
plt.plot(x, y2, color = 'olivedrab', linewidth=1.5)
plt.title('Lineplot', color = 'darkmagenta')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['functie 1', 'functie 2'], bbox_to_anchor=(1,1))
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, sharey=True, sharex=True, figsize = (12, 6))

ax[0,0].plot(x, y1, color = 'darkmagenta', linewidth=1.5, alpha=0.7)
ax[0,0].set_title('functie 1')

ax[0,1].plot(x, y1+y2, color = 'darkmagenta', linewidth=1.5, alpha=0.7)
ax[0,1].set_title('functie 1 + functie 2')

ax[1,0].plot(x, y2, color = 'olivedrab', linewidth=1.5)
ax[1,0].set_title('functie 2')

ax[1,1].plot(x, y1 - y2, color = 'olivedrab', linewidth=1.5)
ax[1,1].set_title('functie 1 - functie 2')

plt.show()
```

```python
n = 200
rx = np.random.randn(n)
ry = np.random.randn(n)
```

```python
plt.grid(linestyle='dashed')
plt.scatter(ry, rx, c='g', s=40, marker='^', alpha=0.8)
plt.scatter(rx, ry, c='maroon', s=40, marker='v', alpha=0.8)
plt.show()
```

```python
fig, ax = plt.subplots(2, 1, sharex = True, figsize = (10, 6))
ax[0].hist(rx, bins=np.arange(-3, 3.5, 0.25), color='maroon')
ax[1].hist(ry, bins=np.arange(-3, 3.5, 0.25), color='green')
plt.show()
```

```python
plt.boxplot([rx, ry], labels=['rx','ry'], widths=0.5, patch_artist = True, boxprops={"facecolor": "gold"})
plt.show()
```

```python
plt.pie([0.2, 0.8], labels=['M','V'], explode=[0.2, 0])
plt.show()
```
