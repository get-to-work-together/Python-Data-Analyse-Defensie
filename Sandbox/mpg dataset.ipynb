## Exploratory Data Analysis

```python
df.info()
```

```python
df.describe()
```

```python
df['origin'].value_counts()
```

```python
origin = {
    1: 'usa',
    2: 'europe',
    3: 'asia'
}

df['origin'] = df['origin'].map(origin)
```

```python
df['car name'].value_counts().sort_index()
```

```python
df[['make', 'model']] = df['car name'].str.split(n=1, expand=True)
```

```python
df
```

```python
df['make'].value_counts().sort_index()
```

```python
fix_typos = {make: make for make in df['make'].unique()}

fix_typos.update({
    'chevy': 'chevrolet',
    'chevroelt': 'chevrolet',
    'maxda': 'mazda',
    'mercedes-benz': 'mercedes',
    'toyouta': 'toyota',
    'vokswagen': 'volkswagen',
    'vw': 'volkswagen'
})

df['make'] = df['make'].map(fix_typos, na_action='ignore')
```

```python
df['make'].value_counts().sort_index()
```
