```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv', sep = ';')
df
```

```python
x = df['Father'].values
y = df['Son'].values
```

```python
plt.scatter(x, y)
plt.xlabel('Fathers')
plt.ylabel('Sons')
plt.show()
```

```python
m = np.corrcoef(x, y)
pearson = m[0, 1]
```

```python
print(f'Pearson correlatie coefficient: {pearson:.3f}')
```
