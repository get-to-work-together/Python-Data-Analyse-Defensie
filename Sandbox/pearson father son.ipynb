```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv', sep = ';')
df
```

```python
x = df['Father'].values
y = df['Son'].values
```

```python
plt.scatter(x, y)
plt.xlabel('Fathers')
plt.ylabel('Sons')
plt.show()
```

```python
m = np.corrcoef(x, y)
pearson = m[0, 1]
```

```python
print(f'Pearson correlatie coefficient: {pearson:.3f}')
```

```python
coefs = np.polyfit(x, y, 1)
coefs
```

```python
model = np.poly1d(coefs)
```

```python
y_predicted = model(x)
```

```python
plt.scatter(x, y)
plt.plot(x, y_predicted, color = 'red')

plt.show()
```
