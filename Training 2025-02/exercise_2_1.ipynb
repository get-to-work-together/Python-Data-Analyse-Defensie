```python
s = np.random.uniform(0, 10, 10000)

plt.hist(s, bins=20)
plt.show()
```

```python
n = 1000

s = np.random.binomial(6, 0.5, n)

values, counts = np.unique(s, return_counts=True)

plt.bar(values, counts/n)
```
