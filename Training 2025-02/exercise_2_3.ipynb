# Exercise 2.3 - Linear Regression

### Linear

```python
a = 2
b = 5

model = lambda x: a * x + b
x_min = 0
x_max = 10
n = 20
```

### Polynomial

```python
a = 3
b = 10
c = 3

model = lambda x: a*x**2 + b*x + c
```

### Exponential

```python
a = 2
b = 1.6
c = 3

model = lambda x: a*b**x + c
```

```python
x_model = np.linspace(x_min, x_max, n)
y_model = model(x_model)

x_data = np.random.uniform(x_min, x_max, n)
spread = 1
noise = np.random.randn(n) * spread
y_data = model(x_data) + noise

plt.plot(x_model, y_model, color='red', linewidth=1, alpha=0.8)
plt.scatter(x_data, y_data)

plt.show()
```

```python
coefs = np.polyfit(x_data, y_data, deg=1)
coefs
```

```python
regression = np.poly1d(coefs)

plt.plot(x_model, y_model, color='red', linewidth=1, alpha=0.3)
plt.scatter(x_data, y_data)
plt.plot(x_model, regression(x_model), color='green', linewidth=2)

plt.show()
```

```python
errors = y_data - regression(x_data)
```

```python
plt.plot(errors)
plt.axhline(0, color='red', linewidth=1)
plt.show()
```

```python
plt.hist(errors)
plt.show()
```

### Correlation

```python
corr = np.corrcoef(x_data, y_data)[1, 0]
corr
```
