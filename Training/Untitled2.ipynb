```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
year = 2020

stations = {
    260: 'De Bilt',
    279: 'Hoogeveen'
}

skiprows = 47

columns_of_interest = ['TG', 'TN', 'TX']
```

```python
filename1 = 'etmgeg_260.txt'    # De Bilt
filename2 = 'etmgeg_279.txt'    # Hoogeveen
```

```python
kwargs = {'skiprows' : 47, 
          'skipinitialspace': True}

df1 = pd.read_csv(filename1, **kwargs)
df2 = pd.read_csv(filename2, **kwargs)

df1.rename(columns = {'# STN': 'STN'}, inplace = True)
df2.rename(columns = {'# STN': 'STN'}, inplace = True)
```

```python
df1.info()
```

```python
df2.info()
```

```python
df1_selection = df1.loc[df1['YYYYMMDD'].astype('str').str.startswith('2020'), 
                        ['STN','YYYYMMDD'] + columns_of_interest]
df2_selection = df2.loc[df2['YYYYMMDD'].astype('str').str.startswith('2020'), 
                        ['STN','YYYYMMDD'] + columns_of_interest]
```

```python
df1_selection.info()
```

```python
df2_selection.info()
```

```python
df = pd.concat([df1_selection, df2_selection])
```

```python
df[['GEMIDDELDE','MINIMUM','MAXIMUM']] = df[['TG','TN','TX']].astype('int') * 0.1
```

```python
df['DATUM'] = pd.to_datetime(df['YYYYMMDD'], format='%Y%m%d')
```

```python
df['STATION'] = df['STN'].astype('int').map(stations)
```

```python
df
```

```python
df = df[['DATUM', 'STATION', 'GEMIDDELDE', 'MINIMUM', 'MAXIMUM']]
```

```python
df
```

```python
df_unstacked = df.set_index(['DATUM', 'STATION']).unstack()
```

```python
df_unstacked = df.pivot(index = 'DATUM', 
                        columns = 'STATION', 
                        values = ['GEMIDDELDE','MINIMUM','MAXIMUM'])
```
