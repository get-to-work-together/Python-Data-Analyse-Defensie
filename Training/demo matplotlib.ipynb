```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
x = np.linspace(1, 10, 101)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x)/(x + 1)
```

```python
plt.title
```

```python
fig, ax = plt.subplots(1, 2, sharey = True, figsize = (8, 4))

ax[0].axhline(0, color = 'k', linewidth = 1)
ax[1].axhline(0, color = 'k', linewidth = 1)

ax[0].plot(x, y1, label = 'Y1')
ax[0].plot(x, y2, label = 'Y2')
ax[1].plot(x, y3, label = 'Y3')

ax[0].grid()
ax[0].legend()
ax[0].set_title('Demo')
ax[0].set_xlabel('X')

ax[1].grid()

plt.show()
```
