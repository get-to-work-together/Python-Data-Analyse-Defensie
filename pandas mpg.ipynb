```python
import pandas as pd
```

```python
filename = 'Opdrachten/datasets/auto-mpg.csv'

df = pd.read_csv(filename, on_bad_lines = 'skip')
```

```python
df
```

```python
df.sort_values('horsepower', ascending=False).head(5)
```

```python
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'car name']
```

```python
df.rename(columns = {'%mpg': 'mpg'})
```

```python
df[['make', 'model']] = df['car name'].str.split(' ', n=1, expand=True)
```

```python
df['make'].value_counts().sort_index()
```

```python
d = {
    'capri':'ford',
    'chevy':'chevrolet',
    'chevroelt':'chevrolet',
    'maxda':'mazda',
    'mercedes-benz':'mercedes',
    'toyouta':'toyota',
    'vokswagen':'volkswagen',
    'vw':'volkswagen',
}

df['make'] = df['make'].replace(d)
```

```python
df['make'].value_counts().sort_index()
```
