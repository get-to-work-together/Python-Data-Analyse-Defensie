```python
import pandas as pd
```

```python
filename = 'Opdrachten/datasets/auto-mpg.csv'

df = pd.read_csv(filename, on_bad_lines = 'skip')
```

```python
df
```

```python
df.sort_values('horsepower', ascending=False).head(5)
```

```python
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'car name']
```

```python
df.rename(columns = {'%mpg': 'mpg'})
```

```python
df[['make', 'model']] = df['car name'].str.split(' ', n=1, expand=True)
```

```python
df['make'].value_counts().sort_index()
```

```python
d = {
    'capri':'ford',
    'chevy':'chevrolet',
    'chevroelt':'chevrolet',
    'maxda':'mazda',
    'mercedes-benz':'mercedes',
    'toyouta':'toyota',
    'vokswagen':'volkswagen',
    'vw':'volkswagen',
}

df['make'] = df['make'].replace(d)
```

```python
origin = {
    1: 'usa',
    2: 'europe',
    3: 'asia'
}

df['origin'] = df['origin'].replace(origin).astype('category')
```

```python
df['make'].value_counts().sort_index()
```

```python
df['make'].value_counts().plot(kind='barh')
```

```python
import seaborn as sns

sns.pairplot(df, hue='origin');
```

```python
df.info()
```

```python
df['duplicated'] = df['make'].duplicated(keep=False)
```

```python
df.drop_duplicates('make', keep='last')
```

```python
df['weight_class'] = pd.cut(df['weight'], [0,2000, 3000, 10000], labels=['light','medium','heavy'])
```

```python
df['make'].str.upper()
```

```python
pd.pivot_table(df, 
               index='origin', 
               columns='weight_class', 
               values='horsepower', 
               aggfunc='count')
```
